```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(z: np.array) -> np.array:
  return np.exp(z) / np.sum(np.exp(z))

def softmax_temperature(z: np.array, T: float=1.0) -> np.array:
  return np.exp(z/T) / np.sum(np.exp(z/T))
```

```python
z = np.array([-3.0, 2.0, -1.0, 10.0, 9.0, 5.0])
probs = softmax(z)
```

```python
plt.bar(range(len(z)), probs)
plt.grid()
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
br1 = np.arange(len(z))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, softmax(z), color ='grey', width = barWidth, edgecolor ='grey', label ='T=1.0')
plt.bar(br2, softmax_temperature(z, 10.0), color ='g', width = barWidth, edgecolor ='grey', label ='T=10.0')
plt.bar(br3, softmax_temperature(z, 0.1), color ='r', width = barWidth, edgecolor ='grey', label ='T=0.1')

plt.legend()
plt.show()
```

```python
z = np.array([5.0, 4.0])
probs = softmax(z)
plt.bar(range(len(z)), probs)
plt.grid()
```
