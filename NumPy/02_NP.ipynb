```python
def get_distances(points: np.array, centers: np.array):
  return np.square(points[:, None, :] - centers).sum(axis=-1)
```

```python
def get_closest_center(dists: np.array):
  return np.argmin(dists, axis=-1)
```

```python
def get_one_hot(closest: np.array, k:  int):
  t = np.eye(k)
  return t[closest, :]
```

```python
def get_new_centers(points: np.array, oh: np.array):
  temp = points[:, None, :] * oh[:, :, None] #[N,  K,  dim]
  return temp.sum(axis=0)/oh.sum(axis=0)[:, None]
```

```python
def mandelbrot_python(xmin, xmax, ymin, ymax, xn, yn, maxiter, horizon=2.0):
    def mandelbrot(z, maxiter):
        c = z
        for n in range(maxiter):
            if abs(z) > horizon:
                return n
            z = z*z + c
        return maxiter
    r1 = [xmin+i*(xmax-xmin)/xn for i in range(xn)]
    r2 = [ymin+i*(ymax-ymin)/yn for i in range(yn)]
    return [mandelbrot(complex(r, i),maxiter) for r in r1 for i in r2]
```

```python

xmin = -1.5
xmax = 0.5
ymin = -1.2
ymax = 1.2
n = 1000
maxiter = 100
horizon = 2.0
xn = n
yn = n
#==============
r1 = np.linspace(xmin, xmax, xn)
r2 = np.linspace(ymin, ymax, yn)
xg, yg = np.meshgrid(r1, r2)
grid = xg + yg*1j
c = grid
ans = np.ones((n,n)) * maxiter

for i in range(maxiter):
  grid = np.where(ans > i, grid*grid + c, grid)
  ans = np.where(np.logical_and(np.abs(grid) < horizon, ans >= i), ans, i)
  # print(grid)
  # print(ans)


plt.imshow(ans, cmap='jet')
```

```python
# Any live cell with fewer than two live neighbours dies, as if by needs caused by underpopulation.
# Any live cell with more than three live neighbours dies, as if by overcrowding.
# Any live cell with two or three live neighbours lives, unchanged, to the next generation.
# Any dead cell with exactly three live neighbours becomes a live cell.
# If s == 3, g = 1. If s == 2 and g == 1, g = 1. Else g = 0

def gameoflife_step(grid: np.array) -> np.array:
  pgrid = np.pad(grid, 1, mode='constant', constant_values=0)
  nbrs = np.lib.stride_tricks.sliding_window_view(pgrid, window_shape=(3,3))
  nbr_sums = nbrs.sum(axis=(-2,-1)) - grid
  return np.logical_or((nbr_sums == 3),np.logical_and(nbr_sums == 2, grid == 1)).astype(np.int64)

def gameoflife_sim(grid: np.array, steps: int=10) -> np.array:
  plt.figure(figsize=(3,3))
  plt.imshow(grid, cmap='gray')
  plt.show()
  for i in range(steps):
    grid = gameoflife_step(grid)
    plt.figure(figsize=(3,3))
    plt.imshow(grid, cmap='gray')
    plt.show()
  return grid
```

```python
class RandomWalker:
  def __init__(self, position: int = 0):
    self.position = position

  def walk(self, n):
    for i in range(n):
      yield self.position
      self.position += random.choice([-1, 1])
```

```python
def random_walk(n: int = 1000, position: int=0):
  walk = [position]
  for i in range(n):
    position += random.choice([-1,1])
    walk.append(position)
  return walk
```

```python
def train_step(points: np.array, centers: np.array):
  assert points.shape[-1] == centers.shape[-1]
  K = centers.shape[0]
  squared_differences = np.square(points - centers[:, None, :]).sum(axis=-1)
  closest_center = np.argmin(squared_differences, axis=0)
  new_centers = np.array([points[closest_center == k].mean(axis=0) for k in range(K)])
  return new_centers, np.sum(squared_differences)
```

```python
num_steps = 5
losses = []
for i in range(num_steps):
  centers, loss = train_step(points, centers)
  losses.append(loss)
  plt.figure(figsize=(5,5))
  plt.scatter(points[:,0], points[:,1])
  plt.scatter(centers[:,0], centers[:,1], c="r", marker="s", s=8)
  plt.show()
plt.plot(losses)
```
